```python
def print_section(title):
    print("\n" + "="*50)
    print(title)
    print("="*50)
```

```python
# 1. Revenue by Product Category
print_section("Revenue by Product Category")
category_revenue_pipeline = [
    # Join orders with order_items
    {"$lookup": {
        "from": "order_items",
        "localField": "product_id",
        "foreignField": "product_id",
        "as": "sales"
    }},
    # Unwind the sales array
    {"$unwind": "$sales"},
    # Group by category and calculate revenue
    {"$group": {
        "_id": "$category",
        "total_revenue": {
            "$sum": {"$multiply": ["$sales.quantity", "$sales.price"]}
        },
        "number_of_sales": {"$sum": 1}
    }},
    # Sort by revenue in descending order
    {"$sort": {"total_revenue": -1}}
]
category_results = list(db.products.aggregate(category_revenue_pipeline))
for result in category_results:
    print(f"Category: {result['_id']}")
    print(f"Total Revenue: ${result['total_revenue']:,.2f}")
    print(f"Number of Sales: {result['number_of_sales']}\n")
```

```python
# 2. Average Delivery Time
print_section("Average Delivery Time by Status")
delivery_time_pipeline = [
    # Match only delivered orders
    {"$match": {"status": "Delivered"}},
    # Calculate average time since order date
    {"$group": {
        "_id": None,
        "avg_delivery_time_days": {
            "$avg": {
                "$divide": [
                    {"$subtract": [datetime.now(), "$order_date"]},
                    1000 * 60 * 60 * 24  # Convert milliseconds to days
                ]
            }
        },
        "total_delivered_orders": {"$sum": 1}
    }}
]

delivery_results = list(db.orders.aggregate(delivery_time_pipeline))
if delivery_results:
    print(f"Average Delivery Time: {delivery_results[0]['avg_delivery_time_days']:.1f} days")
    print(f"Total Delivered Orders: {delivery_results[0]['total_delivered_orders']}")
```

```python
# 3. Customers by State
print_section("Customer Distribution by State")
state_distribution_pipeline = [
    # Group by state
    {"$group": {
        "_id": "$address.state",
        "customer_count": {"$sum": 1}
    }},
    # Sort by customer count in descending order
    {"$sort": {"customer_count": -1}}
]

state_results = list(db.customers.aggregate(state_distribution_pipeline))
for result in state_results:
    print(f"State: {result['_id']}")
    print(f"Number of Customers: {result['customer_count']}\n")
```

```python
# 4. Top 3 Most Expensive Products per Order
print_section("Top 3 Most Expensive Products per Order")
expensive_products_pipeline = [
    # Join with products collection
    {"$lookup": {
        "from": "products",
        "localField": "product_id",
        "foreignField": "product_id",
        "as": "product_info"
    }},
    # Unwind the product_info array
    {"$unwind": "$product_info"},
    # Group by order_id
    {"$group": {
        "_id": "$order_id",
        "products": {
            "$push": {
                "product_name": "$product_info.product_name",
                "price": "$price",
                "quantity": "$quantity",
                "total": {"$multiply": ["$price", "$quantity"]}
            }
        }
    }},
    # Sort products within each group
    {"$project": {
        "top_products": {
            "$slice": [
                {"$sortArray": {
                    "input": "$products",
                    "sortBy": {"total": -1}
                }},
                3
            ]
        }
    }},
    # Limit to first 5 orders for display
    {"$limit": 5}
]

expensive_results = list(db.order_items.aggregate(expensive_products_pipeline))
for result in expensive_results:
    print(f"\nOrder ID: {result['_id']}")
    for idx, product in enumerate(result['top_products'], 1):
        print(f"{idx}. {product['product_name']}")
        print(f"   Price: ${product['price']:.2f}")
        print(f"   Quantity: {product['quantity']}")
        print(f"   Total: ${product['total']:.2f}")
```

```python

# Schema Design Implementation
"""
Schema Design Decisions:

1. Customers Collection (Referenced)
{
    customer_id: int,
    name: string,
    email: string,
    address: {
        street: string,
        city: string,
        state: string
    }
}
Rationale: 
- Customer data is accessed independently of orders
- Address is embedded within customer as it's integral to customer data
- Customer data is relatively static and queried frequently

2. Products Collection (Referenced)
{
    product_id: int,
    product_name: string,
    category: string,
    price: decimal
}
Rationale:
- Products are queried independently for catalog browsing
- Product data is accessed frequently across different orders
- Product information updates should reflect across all orders

3. Orders Collection (Referenced)
{
    order_id: int,
    customer_id: int,  // Reference to customers collection
    order_date: date,
    status: string
}
Rationale:
- Orders can be queried independently
- Multiple orders per customer requires flexibility
- Allows for efficient querying of order history

4. Order_Items Collection (Referenced)
{
    order_item_id: int,
    order_id: int,     // Reference to orders collection
    product_id: int,   // Reference to products collection
    quantity: int,
    price: decimal
}
Rationale:
- Separating order items allows for efficient order processing
- Enables easy querying of product purchase history
- Maintains data consistency for price changes
"""
```

```python
from pymongo import MongoClient, ASCENDING, DESCENDING
from pymongo.errors import DuplicateKeyError

def check_and_fix_duplicates():
    """Check for and report any duplicate values in key fields before creating unique indexes."""
    
    # Check products collection
    product_duplicates = db.products.aggregate([
        {"$group": {
            "_id": "$product_id",
            "count": {"$sum": 1},
            "docs": {"$push": "$_id"}
        }},
        {"$match": {
            "count": {"$gt": 1}
        }}
    ])
    
    # Print duplicate products if any
    duplicates_found = False
    for dup in product_duplicates:
        duplicates_found = True
        print(f"Duplicate product_id found: {dup['_id']} appears {dup['count']} times")
    
    if duplicates_found:
        user_input = input("Would you like to automatically fix duplicates by updating product_ids? (y/n): ")
        if user_input.lower() == 'y':
            fix_duplicate_products()
        else:
            print("Please fix the duplicate product_ids manually before creating indexes.")
            return False
    
    return True

def fix_duplicate_products():
    """Fix duplicate product_ids by assigning new unique IDs."""
    
    # Get the highest existing product_id
    max_product = db.products.find_one(sort=[("product_id", -1)])
    next_id = (max_product["product_id"] + 1) if max_product else 1000
    
    # Find and fix duplicates
    seen_product_ids = set()
    
    for product in db.products.find():
        if product["product_id"] in seen_product_ids:
            # Update to new unique ID
            db.products.update_one(
                {"_id": product["_id"]},
                {"$set": {"product_id": next_id}}
            )
            print(f"Updated duplicate product_id {product['product_id']} to {next_id}")
            next_id += 1
        else:
            seen_product_ids.add(product["product_id"])

```

```python
# Create Indexes
def create_indexes():
    # Customers Collection Indexes
    db.customers.create_index([("customer_id", ASCENDING)], unique=True)
    db.customers.create_index([("email", ASCENDING)], unique=True)
    db.customers.create_index([("address.state", ASCENDING)])  # For geographic queries

    # Products Collection Indexes
    db.products.create_index([("product_id", ASCENDING)], unique=True)
    db.products.create_index([("category", ASCENDING)])  # For category-based queries
    db.products.create_index([("product_name", ASCENDING)])  # For product search

    # Orders Collection Indexes
    db.orders.create_index([("order_id", ASCENDING)], unique=True)
    db.orders.create_index([("customer_id", ASCENDING)])  # For customer order history
    db.orders.create_index([("order_date", DESCENDING)])  # For date-based queries
    db.orders.create_index([("status", ASCENDING)])  # For status-based queries

    # Order_Items Collection Indexes
    db.order_items.create_index([("order_item_id", ASCENDING)], unique=True)
    db.order_items.create_index([("order_id", ASCENDING)])  # For order details
    db.order_items.create_index([("product_id", ASCENDING)])  # For product sales analysis

    # Compound Indexes for Common Query Patterns
    db.orders.create_index([("customer_id", ASCENDING), ("order_date", DESCENDING)])  # For customer order history with date
    db.order_items.create_index([("order_id", ASCENDING), ("product_id", ASCENDING)])  # For order item lookups

def print_indexes():
    print("\nCurrent Indexes in Collections:")
    collections = ["customers", "products", "orders", "order_items"]
    
    for collection in collections:
        print(f"\n{collection.upper()} Indexes:")
        indexes = db[collection].list_indexes()
        for index in indexes:
            print(f"- {index['name']}: {index['key']}")

# Execute index creation and print results
create_indexes()
print_indexes()

# Verification queries to test indexes
def verify_indexes():
    print("\nVerifying Index Usage:")
    
    # Test customer lookup
    explain_result = db.customers.find({"customer_id": 1}).explain()
    print("\nCustomer Lookup by ID:")
    print(f"Index Used: {explain_result.get('queryPlanner', {}).get('winningPlan', {}).get('inputStage', {}).get('indexName', 'No index used')}")
    
    # Test order history lookup
    explain_result = db.orders.find({"customer_id": 1}).sort("order_date", -1).explain()
    print("\nCustomer Order History:")
    print(f"Index Used: {explain_result.get('queryPlanner', {}).get('winningPlan', {}).get('inputStage', {}).get('indexName', 'No index used')}")

verify_indexes()
```

```python
def get_next_sequence(collection_name, id_field):
    """Get the next available ID for a given collection."""
    last_doc = db[collection_name].find_one(
        sort=[(id_field, -1)]
    )
    if last_doc and id_field in last_doc:
        return last_doc[id_field] + 1
    return 1

def create_order_with_inventory_update(customer_id, items):
    """
    Create an order and update inventory atomically using transactions.
    
    Args:
        customer_id: ID of the customer placing the order
        items: List of dictionaries containing product_id and quantity
    """
    # Start a session for the transaction
    with client.start_session() as session:
        try:
            # Start transaction
            with session.start_transaction():
                # Get next available order_id
                next_order_id = get_next_sequence("orders", "order_id")
                
                # Create new order
                order = {
                    "order_id": next_order_id,
                    "customer_id": customer_id,
                    "order_date": datetime.now(),
                    "status": "Pending"
                }
                db.orders.insert_one(order, session=session)
                
                # Create order items
                for item in items:
                    # Check inventory
                    product = db.products.find_one(
                        {"product_id": item["product_id"]},
                        session=session
                    )
                    if not product:
                        raise ValueError(f"Product {item['product_id']} not found")
                    
                    # Get next available order_item_id
                    next_item_id = get_next_sequence("order_items", "order_item_id")
                    
                    # Create order item
                    order_item = {
                        "order_item_id": next_item_id,
                        "order_id": next_order_id,
                        "product_id": item["product_id"],
                        "quantity": item["quantity"],
                        "price": product["price"]
                    }
                    db.order_items.insert_one(order_item, session=session)
                
                print(f"Order {order['order_id']} created successfully")
                return order['order_id']
                
        except PyMongoError as e:
            print(f"Transaction failed: {str(e)}")
            raise

def monitor_orders():
    """Monitor real-time changes in the orders collection."""
    try:
        pipeline = [
            {"$match": {
                "operationType": {"$in": ["insert", "update", "delete"]}
            }}
        ]
        
        with db.orders.watch(pipeline) as stream:
            print("Monitoring orders collection for changes... Press Ctrl+C to stop.")
            while True:
                try:
                    change = stream.next()
                    print("\nChange detected:")
                    print(f"Operation: {change['operationType']}")
                    
                    if change['operationType'] == 'insert':
                        print(f"New order: {change['fullDocument']}")
                    elif change['operationType'] == 'update':
                        print(f"Updated order ID: {change['documentKey']['_id']}")
                        print(f"Updated fields: {change['updateDescription']['updatedFields']}")
                    elif change['operationType'] == 'delete':
                        print(f"Deleted order ID: {change['documentKey']['_id']}")
                        
                except KeyboardInterrupt:
                    break
                
    except PyMongoError as e:
        print(f"Error monitoring changes: {str(e)}")

def setup_product_validation():
    """Set up schema validation for the products collection."""
    try:
        validator = {
            "$jsonSchema": {
                "bsonType": "object",
                "required": ["product_id", "product_name", "category", "price"],
                "properties": {
                    "product_id": {
                        "bsonType": "int",
                        "description": "must be an integer and is required"
                    },
                    "product_name": {
                        "bsonType": "string",
                        "description": "must be a string and is required"
                    },
                    "category": {
                        "bsonType": "string",
                        "description": "must be a string and is required"
                    },
                    "price": {
                        "bsonType": "double",
                        "minimum": 0.01,
                        "description": "must be a positive number and is required"
                    }
                }
            }
        }
        
        db.command("collMod", "products", 
                   validator=validator,
                   validationLevel="strict",
                   validationAction="error")
        print("Product schema validation set up successfully")
        
    except PyMongoError as e:
        print(f"Error setting up schema validation: {str(e)}")

def test_features():
    """Test the advanced features implementation."""
    try:
        # Test transaction
        test_order = {
            "customer_id": 1,
            "items": [
                {"product_id": 101, "quantity": 1},
                {"product_id": 102, "quantity": 2}
            ]
        }
        
        print("\nTesting transaction...")
        order_id = create_order_with_inventory_update(
            test_order["customer_id"],
            test_order["items"]
        )
        print(f"Created order with ID: {order_id}")
        
        # Set up schema validation
        print("\nSetting up schema validation...")
        setup_product_validation()
        
        # Test schema validation
        print("\nTesting schema validation...")
        try:
            # This should fail due to missing price
            invalid_product = {
                "product_id": 999,
                "product_name": "Test Product",
                "category": "Test"
            }
            db.products.insert_one(invalid_product)
        except PyMongoError as e:
            print(f"Validation caught invalid product: {str(e)}")
            
    except Exception as e:
        print(f"Error in test_features: {str(e)}")

if __name__ == "__main__":
    test_features()
    # Uncomment to monitor changes:
    # monitor_orders()
```
